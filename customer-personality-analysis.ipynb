## Data cleaning

The Income variable has some missing values. since the percentage is very low. I will remove themdf1=na.omit(df)

```r
missing.values <- df %>%
  gather(key = "key", value = "val") %>%
  mutate(isna = is.na(val)) %>%
  group_by(key) %>%
  mutate(total = n()) %>%
  group_by(key, total, isna) %>%
  summarise(num.isna = n(), .groups = 'drop') %>%
  mutate(pct = num.isna / total * 100) 

levels =(missing.values  %>% filter(isna == T) %>% arrange(desc(pct)))$key


d <- missing.values %>%
     filter(key == levels)

pie_chart <- ggplot(data = d, aes(x = "", y = pct, fill = isna)) +
  geom_bar(width = 1, stat = "identity") +
  coord_polar(theta = "y") +
  scale_fill_manual(values = c("#ffb997", "red"), labels = c("Present", "Missing")) +
  labs(title = "Percentage of missing values", fill = "") +
  geom_text(aes(label = paste(round(pct, 2), "%")), position = position_stack(vjust = 0.5)) +
  my_theme()

pie_chart
```

```r
# convert categorical variables to numerical values
customer_data <- df %>% 
        mutate( Education = if_else(Education %in% c("2n Cycle", "Basic", "Graduation"),
                             "Undergraduate",
                             "Postgraduate"),
              Marital_Status = if_else(Marital_Status %in% c("Together", "Married"),
                             "In relationship",
                             Marital_Status),
               Marital_Status = if_else(Marital_Status %in% c("Alone", "Single"),
                             "Single",
                             Marital_Status),
              age = year(Sys.Date()) - Year_Birth)
customer_data <- customer_data %>% 
  select(-ID, -Year_Birth, -Z_CostContact, -Z_Revenue, -Complain) %>%
  subset( Marital_Status != "YOLO")%>%
  mutate(age_group = cut(as.numeric(age), breaks = c(25, 35, 45, 55, 65, 75, Inf),
                    labels = c("Under 35", "35-45", "46-55", "56-65", "66-75", "Over 75")))
head(customer_data,4)
```
